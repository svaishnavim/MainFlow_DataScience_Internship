```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("householdtask3.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

## Bar Chart

```python
plt.figure(figsize=(10, 6))

# Bar chart for 'own' and 'own_wm'
plt.subplot(1, 2, 1)
plt.bar(df['year'], df['own'], label='Own', color='b', alpha=0.6)
plt.bar(df['year'], df['own_wm'], label='Own WM', color='r', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Ownership and Ownership with WM by Year')
plt.legend()
plt.show()
```

## Line Chart

```python
plt.subplot(1, 2, 2)
plt.plot(df['year'], df['income'], label='Income', color='g')
plt.plot(df['year'], df['expenditure'], label='Expenditure', color='m')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Income and Expenditure by Year')
plt.legend()
plt.show()
```
